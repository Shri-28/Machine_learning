```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
from sklearn.model_selection import train_test_split
```

```python
tou = 0.7
#X_TRAIN DATA
X_train = np.array(list(range(3,33))+[3.2,6.2])
X_train = X_train[:,np.newaxis]
print("X_TRAIN:\n",X_train)

#Y_TRAIN DATA
y_train = np.array([1,2,1,2,1,1,3,4,5,4,5,6,5,6,7,8,9,10,11,11,12,11,11,10,12,11,11,10,9,8,2,13])
print("Y_TRAIN:\n",y_train)

#X_TEST DATA
X_test = np.array([i/10. for i in range(400)])
X_test = X_test[:,np.newaxis]
print("X_TEST:\n",y_train)

#Y_TEST DATA
y_test = []
print("Y_TEST:\n",y_test)

```

```python
#FIND EUCLIDEAN DISTANCE B/T TRAIN AND TEST POINTS
count = 0
for r in range(len(X_test)):
    try:
        wts = np.exp(-np.sum((X_train-X_test[r])**2,axis=1)/(2*tou**2))
        W = np.diag(wts)
        factor1 = np.linalg.inv(X_train.T.dot(W).dot(X_train))
        parameters = factor1.dot(X_train.T).dot(W).dot(y_train)
        prediction = X_test[r].dot(parameters)
        y_test.append(prediction)
        count = count + 1
    except:
        pass
y_test = np.array(y_test)
print("Y_TEST:\n",y_test)
```

```python
#DRAW THE GRAPH
pt.plot(X_train.squeeze(),y_train,"o")
pt.plot(X_test.squeeze(),y_test,"*")
pt.show()
```
